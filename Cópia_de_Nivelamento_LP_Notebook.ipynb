##4Ô∏è‚É£ **Estrutura `match case`**

O bloco `match case` √© uma estrutura de controle de fluxo que permite comparar uma vari√°vel com diferentes valores ou padr√µes de forma mais organizada e leg√≠vel do que as tradicionais estruturas `if/elif/else`.

**üìå Quando usar?**

- Quando h√° m√∫ltiplas condi√ß√µes que precisam ser verificadas.

 **üìå Sintaxe:**

```python
opcao = 'a'

match opcao:
    case 'a':
        print("Op√ß√£o A selecionada")
    case 'b':
        print("Op√ß√£o B selecionada")
    case _: #caso padr√£o, se n√£o ouver op√ß√£o v√°lida (funciona como um else)
        print("Op√ß√£o inv√°lida")
```





 **üìå Exemplo:**
 Comparando Strings

````python
nome = "Jorge"

match nome:
    case "Jorge":
        print("Ol√°, Jorge!")
    case "Maria":
        print("Ol√°, Maria!")
    case _:
        print("N√£o reconhe√ßo voc√™!")
````

 **üìå Exemplo:**
 Dias da semana

````python
dia = int(input("Digite um n√∫mero entre 1 e 7: "))

match dia:
    case 1:
        print("Domingo")
    case 2:
        print("Segunda-feira")
    case 3:
        print("Ter√ßa-feira")
    case 4:
        print("Quarta-feira")
    case 5:
        print("Quinta-feira")
    case 6:
        print("Sexta-feira")
    case 7:
        print("S√°bado")
    case _:
        print("N√∫mero inv√°lido")
````

---


## **1Ô∏è‚É£ O que s√£o Estruturas de Repeti√ß√£o?**


üí° **Defini√ß√£o:**  
Uma estrutura de repeti√ß√£o permite que um programa **repita um bloco de c√≥digo v√°rias vezes** enquanto uma condi√ß√£o for verdadeira.  



üìç **Exemplo na vida real:**

- üö¥ Pedalar at√© chegar ao destino.  
- üçΩ Comer at√© ficar satisfeito.  
- ‚è∞ Repetir o despertador at√© acordar.  



üìå **Tipos de la√ßos em Python:**  

- `while` ‚Üí Usa uma **condi√ß√£o** para continuar repetindo.  
- `for` ‚Üí Usa um **contador** para repetir um n√∫mero definido de vezes (veremos na pr√≥xima aula).

## **2Ô∏è‚É£ Estrutura WHILE**



 üìå **Como funciona?**  
1Ô∏è‚É£ O programa verifica a condi√ß√£o.  
2Ô∏è‚É£ Se for **verdadeira**, executa o bloco de c√≥digo.  
3Ô∏è‚É£ Volta para o passo 1 at√© que a condi√ß√£o seja falsa.  




üìå **Sintaxe:**  

```python
while condi√ß√£o:
    # c√≥digo a ser repetido
```



 üìå **Exemplo 1:**

Contagem de 1 a 5  


````python
contador = 1

while contador <= 5:
    print("N√∫mero:", contador)
    contador += 1  # Incremento para evitar loop infinito
````


 üìå **Exemplo 2:**

Usu√°rio s√≥ sai quando digita `"sair"`  

````python
comando = ""

while comando.lower() != "sair":
    comando = input("Digite um comando (ou 'sair' para encerrar): ")
    print("Voc√™ digitou:", comando)

print("Programa encerrado.")
````

## **3Ô∏è‚É£ Cuidados com o WHILE (Loops Infinitos!)**



‚ö† **Se n√£o atualizarmos a vari√°vel da condi√ß√£o, o loop nunca para!**  

```python
# ‚ö†Ô∏è C√ìDIGO ERRADO! Vai rodar infinitamente!

    # Esquecemos de incrementar x, ent√£o o loop nunca termina!
```

## **4Ô∏è‚É£ Simulando DO-WHILE em Python**



üìå **O `do-while` existe em outras linguagens, mas Python n√£o tem essa estrutura nativa.**  
üìå **Podemos simular usando `while True` e `break`.**  

üìå **Exemplo: O usu√°rio deve digitar um n√∫mero maior que 10 para sair.**  


````python
while True:
    numero = int(input("Digite um n√∫mero maior que 10: "))

    if numero > 10:
        print("N√∫mero v√°lido! Saindo do loop.")
        break  # Sai do loop quando a condi√ß√£o √© satisfeita
    else:
        print("N√∫mero inv√°lido. Tente novamente.")
````

 üéØ **Desafio:**  
- Criar um loop que pede notas ao usu√°rio at√© que ele digite `-1`.  
- Depois, exibir a **m√©dia das notas** digitadas.  

üí° **Dicas:**  
- Criar uma vari√°vel `soma` para armazenar as notas.  
- Criar uma vari√°vel `contador` para contar quantas notas foram inseridas.

**üì§ Exemplo de Entrada e Sa√≠da:**
- Entrada do usu√°rio:



```
Digite uma nota (ou -1 para encerrar): 7.5  
Digite uma nota (ou -1 para encerrar): 8.0  
Digite uma nota (ou -1 para encerrar): 6.5  
Digite uma nota (ou -1 para encerrar): -1  
```



- Sa√≠da esperada:



```
A m√©dia das notas digitadas √©: 7.33
```






````python
#SUA SOLU√á√ÉO
````

##### **üìå Solu√ß√£o:**

````python
# Inicializando vari√°veis
soma = 0  # Armazena a soma das notas
contador = 0  # Conta quantas notas foram inseridas

# Loop para receber notas
while True:
    nota = float(input("Digite uma nota (ou -1 para encerrar): "))

    if nota == -1:  # Condi√ß√£o de sa√≠da do loop
        break

    soma += nota  # Soma as notas digitadas
    contador += 1  # Incrementa o contador

# Calculando e exibindo a m√©dia
if contador > 0:
    media = soma / contador
    print(f"\nA m√©dia das notas digitadas √©: {media:.2f}")
else:
    print("\nNenhuma nota foi digitada.")
````

**üìù Explica√ß√£o**
1. Inicializa√ß√£o de vari√°veis:

    - `soma`: para armazenar a soma das notas.
    - `contador`: para contar quantas notas foram digitadas.
2. Loop `while True`:

    - O programa continua pedindo notas ao usu√°rio.
    - Se o usu√°rio digitar `-1`, o loop √© encerrado com `break`.
3. C√°lculo da m√©dia:

    - Se pelo menos uma nota foi digitada `(contador > 0)`, a m√©dia √© calculada e exibida.
    - Caso contr√°rio, exibe `"Nenhuma nota foi digitada."` para evitar divis√£o por zero.

## **3Ô∏è‚É£ Sistema de Notas da EST/UEA**  

üìå **Descri√ß√£o:**  
O programa deve calcular a **m√©dia parcial (MP)** das avalia√ß√µes de um aluno e verificar a situa√ß√£o dele conforme as regras da **EST/UEA**.  


- O usu√°rio informa quantas **Avalia√ß√µes Parciais (AP)** ser√£o realizadas.  
- O programa calcula a m√©dia das notas e verifica:  
  - **Aprovado direto:** se a m√©dia parcial for maior ou igual a **8.0**.  
  - **Reprovado direto:** se a m√©dia parcial for menor que **4.0**.  
  - **Prova Final (PF):** se a m√©dia parcial estiver entre **4.0 e 7.9**. Nesse caso, a m√©dia final √© calculada como:  

  ```
  MF = (2 * MP + PF) / 3
  ```

  O aluno precisa de uma **MF ‚â• 6.0** para ser aprovado. O programa tamb√©m deve mostrar a **nota m√≠nima necess√°ria na PF** para o aluno ser aprovado.  



---

 **Exemplo de Entrada e Sa√≠da**  



**Exemplo 1: Aprovado Direto**  
```
Entrada:  
Quantas avalia√ß√µes parciais? 2  
Digite a nota da AP1: 9  
Digite a nota da AP2: 8  

Sa√≠da:  
M√©dia Parcial: 8.5  
Parab√©ns! Voc√™ foi aprovado direto.  
```

 **Exemplo 2: Reprovado Direto**  
```
Entrada:  
Quantas avalia√ß√µes parciais? 3  
Digite a nota da AP1: 3  
Digite a nota da AP2: 2  
Digite a nota da AP3: 3  

Sa√≠da:  
M√©dia Parcial: 2.67  
Infelizmente, voc√™ foi reprovado direto.  
```

 **Exemplo 3: Prova Final Necess√°ria**  
```
Entrada:
Quantas avalia√ß√µes parciais? 3
Digite a nota da AP1: 5
Digite a nota da AP2: 1.2
Digite a nota da AP3: 7

Sa√≠da:
M√©dia Parcial: 4.40
Voc√™ precisa fazer a prova final.

Sa√≠da:
M√©dia Final: 5.93
Infelizmente, voc√™ foi reprovado.
```



````python
# SEU C√ìDIGO
````

##### **üìå Solu√ß√£o 3:**

````python
# Entrada da quantidade de avalia√ß√µes
n_avaliacoes = int(input("Quantas avalia√ß√µes parciais? "))

# Entrada das notas das avalia√ß√µes
notas = []
for i in range(n_avaliacoes):
    nota = float(input(f"Digite a nota da AP{i}: "))
    notas.append(nota)

# Calculando a m√©dia parcial
media_parcial = sum(notas) / n_avaliacoes
print(f"\nM√©dia Parcial: {media_parcial:.2f}")

# Verificando a situa√ß√£o do aluno
if media_parcial >= 8.0:
    print("Parab√©ns! Voc√™ foi aprovado direto.")
elif media_parcial < 4.0:
    print("Infelizmente, voc√™ foi reprovado direto.")
else:
    print("Voc√™ precisa fazer a prova final.")

    # Calculando a nota m√≠nima necess√°ria na prova final
    nota_minima_pf = 3 * 6 - 2 * media_parcial
    print(f"Para ser aprovado, voc√™ precisa tirar no m√≠nimo {nota_minima_pf:.2f} na prova final.")

    # Entrada da nota da prova final
    prova_final = float(input("\nDigite a nota da Prova Final: "))

    # Calculando a m√©dia final
    media_final = (2 * media_parcial + prova_final) / 3
    print(f"M√©dia Final: {media_final:.2f}")

    # Verificando a situa√ß√£o final
    if media_final >= 6.0:
        print("Parab√©ns! Voc√™ foi aprovado na prova final.")
    else:
        print("Infelizmente, voc√™ foi reprovado.")
````

**üìù Explica√ß√£o Passo a Passo:**  

1. **Entrada das Avalia√ß√µes:**  
   O usu√°rio informa o n√∫mero de avalia√ß√µes (`n_avaliacoes`).  
   Um la√ßo `for` √© usado para coletar a nota de cada **Avalia√ß√£o Parcial (AP)**, armazenando todas as notas em uma lista `notas`.  

2. **C√°lculo da M√©dia Parcial (MP):**  
   A m√©dia parcial √© calculada usando `sum(notas) / n_avaliacoes`.  

3. **Verifica√ß√£o da Situa√ß√£o Inicial:**  
   - **Aprovado direto:** se `MP ‚â• 8.0`.  
   - **Reprovado direto:** se `MP < 4.0`.  
   - **Prova Final:** se `4.0 ‚â§ MP < 8.0`.  

4. **C√°lculo da Nota M√≠nima para a Prova Final:**  
   A nota m√≠nima necess√°ria na prova final para obter m√©dia final (MF) igual a **6.0** √© calculada com a f√≥rmula:  
   ```
   Nota M√≠nima PF = 3 * 6 - 2 * MP
   ```  

5. **Verifica√ß√£o Final:**  
   - O usu√°rio digita a nota da **Prova Final (PF)**.  
   - A **M√©dia Final (MF)** √© calculada como `(2 * MP + PF) / 3`.  
   - Se `MF ‚â• 6.0`, o aluno √© aprovado; caso contr√°rio, √© reprovado.
